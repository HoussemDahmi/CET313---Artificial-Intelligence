```python
import time

# BFS algorithm
def bfs(graph, start, goal):
    visited = []  # Visited nodes list
    queue = [(start, [])]  # Queue list
    while queue:
        node, path = queue.pop(0)  # Queue the first element
        if node == goal:
            return path + [node]
        if node not in visited:
            visited.append(node)
            for neighbour in graph[node]:
                queue.append((neighbour, path + [node]))
    return None  # Goal not found

# DFS algorithm
def dfs(graph, start, goal):
    visited = []  # Visited nodes list
    stack = [(start, [])]  # Stack for DFS (similar to the queue in BFS)
    while stack:
        node, path = stack.pop()  # Pop the last element (LIFO for DFS)
        if node == goal:
            return path + [node]
        if node not in visited:
            visited.append(node)
            # Iterate through neighbours in reverse to maintain order like recursion
            for neighbour in reversed(graph[node]):
                stack.append((neighbour, path + [node]))
    return None  # Goal not found

graph1 = {
    'A': ['B', 'E'],
    'B': ['C', 'D'],
    'C': ['C'],
    'D': ['D'],
    'E': ['F', 'G'],
    'F': ['F'],
    'G': ['G']
}

graph2 = {
    'A': ['B', 'C'], 
    'B': ['D', 'E'], 
    'C': ['F'], 
    'D': ['G'], 
    'E': [],
    'F': ['E'],
    'G': ['H'],
    'H': ['F']
}

# start and goal nodes for both graphs
start_node = 'A'
goal_node = 'G'

# Measure and print the time for BFS and DFS on graph1
print("Testing on Graph 1")
start_time = time.time()
bfs_result = bfs(graph1, start_node, goal_node)
end_time = time.time()
print(f"BFS Time: {end_time - start_time:.6f} seconds, Result: {bfs_result}")

start_time = time.time()
dfs_result = dfs(graph1, start_node, goal_node)
end_time = time.time()
print(f"DFS Time: {end_time - start_time:.6f} seconds, Result: {dfs_result}")

# Measure and print the time for BFS and DFS on graph2
print("\nTesting on Graph 2")
start_time = time.time()
bfs_result = bfs(graph2, start_node, goal_node)
end_time = time.time()
print(f"BFS Time: {end_time - start_time:.6f} seconds, Result: {bfs_result}")

start_time = time.time()
dfs_result = dfs(graph2, start_node, goal_node)
end_time = time.time()
print(f"DFS Time: {end_time - start_time:.6f} seconds, Result: {dfs_result}")
```

```python
# Graph definition
graph = {
    'A': ['B', 'C'], 
    'B': ['D', 'E'], 
    'C': ['F'], 
    'D': ['G'], 
    'E': [],
    'F': ['E'],
    'G': ['H'],
    'H': ['F']
}

# dls algorithm
def dls(graph, start, goal, limit):
    visited = []  # Visited nodes list
    stack = [(start, [], 0)]  # Stack for DFS with (node, path, depth)
    while stack:
        node, path, depth = stack.pop()
        # Display the current node, path, and depth
        print(f"Current node: {node}, Path: {path + [node]}, Depth: {depth}")
        
        if node == goal:
            return path + [node]
        
        if node not in visited:
            visited.append(node)
            if depth < limit:  # Only expand if depth is below the limit
                # Iterate through neighbours in reverse to maintain DFS behavior
                for neighbour in reversed(graph[node]):
                    stack.append((neighbour, path + [node], depth + 1))
    
    print(f"Goal not found within depth limit {limit}")
    return None  # Goal not found within the depth limit

# Define start and goal nodes
start_node = 'D'
goal_node = 'F'

# Initial depth limit
max_depth_limit = 5

# Test DLS with increasing depth limits
for limit in range(1, max_depth_limit + 1):
    print(f"\nTesting with depth limit: {limit}")
    result = dls(graph, start_node, goal_node, limit)
    if result:
        print("Path found:", result)
        break
    else:
        print("No path found within this depth limit.")
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H', 'I'],
    'E': ['J'],
    'F': ['K', 'L'],
    'G': ['M'],
    'H': [],
    'I': [],
    'J': [],
    'K': [],
    'L': [],
    'M': []
}

# Define BFS algorithm
def bfs(graph, start, goal):
    visited = []  # Visited nodes list
    queue = [(start, [])]  # Queue list
    while queue:
        node, path = queue.pop(0)  # Queue the first element
        if node == goal:
            return path + [node]
        if node not in visited:
            visited.append(node)
            for neighbour in graph[node]:
                queue.append((neighbour, path + [node]))
    return None  # Goal not found

# Define DFS algorithm
def dfs(graph, start, goal):
    visited = []  # Visited nodes list
    stack = [(start, [])]  # Stack for DFS (similar to the queue in BFS)
    while stack:
        node, path = stack.pop()  # Pop the last element (LIFO for DFS)
        if node == goal:
            return path + [node]
        if node not in visited:
            visited.append(node)
            # Iterate through neighbours in reverse to maintain order like recursion
            for neighbour in reversed(graph[node]):
                stack.append((neighbour, path + [node]))
    return None  # Goal not found
# Depth-Limited Search (DLS) algorithm
def dls(graph, start, goal, limit):
    visited = []  # Visited nodes list
    stack = [(start, [], 0)]  # Stack for DFS with (node, path, depth)
    while stack:
        node, path, depth = stack.pop()
        # Display the current node, path, and depth
        print(f"Current node: {node}, Path: {path + [node]}, Depth: {depth}")
        
        if node == goal:
            return path + [node]
        
        if node not in visited:
            visited.append(node)
            if depth < limit:  # Only expand if depth is below the limit
                # Iterate through neighbours in reverse to maintain DFS behavior
                for neighbour in reversed(graph[node]):
                    stack.append((neighbour, path + [node], depth + 1))
    
    print(f"Goal not found within depth limit {limit}")
    return None  # Goal not found within the depth limit

#defining start & gaol nodes
start_node = 'A'
goal_node = 'H'

# Running BFS
print("BFS Result:")
result_bfs = bfs(graph, start_node, goal_node)
print("Path found using BFS:", result_bfs)

# Running DFS
print("\nDFS Result:")
result_dfs = dfs(graph, start_node, goal_node)
print("Path found using DFS:", result_dfs)

# Running DLS with a depth limit 4
depth_limit = 4  
print(f"\nDLS Result with depth limit {depth_limit}:")
result_dls = dls(graph, start_node, goal_node, depth_limit)
print("Path found using DLS:", result_dls)
```
